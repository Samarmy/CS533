```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
def getArrayArango(var1):
    y = []
    with open('/s/chopin/k/grad/sarmst/cs533/graphs/arango/'+ var1 +'_times','r') as csvfile:
        plots = csv.reader(csvfile, delimiter=',')
        for row in plots:
            y.append(float(row[0]))

    return y[1:]

def getArrayNeo4j(var1):
    y = []
    with open('/s/chopin/k/grad/sarmst/cs533/graphs/neo4j/'+ var1 +'_times','r') as csvfile:
        plots = csv.reader(csvfile, delimiter=',')
        for row in plots:
            y.append(float(row[0]))

    return y[1:]

def avg(lst): 
    return sum(lst) / len(lst) 

def subtractLists(list1, list2):
    listFinal= []
    for x in range(0, len(list1)):
        listFinal.append(list1[x] - list2[x])
    return listFinal

def divideLists(list1, list2):
    listFinal= []
    for x in range(0, len(list1)):
        listFinal.append((list1[x] / list2[x])*100)
    return listFinal
```

```python
wiki_neo4j_sum = avg(getArrayNeo4j('sum1'))/1000
wiki_neo4j_avg = avg(getArrayNeo4j('avg1'))/1000
wiki_neo4j_count = avg(getArrayNeo4j('count1'))/1000
wiki_neo4j_min = avg(getArrayNeo4j('min1'))/1000
wiki_neo4j_max = avg(getArrayNeo4j('max1'))/1000
wiki_neo4j_kneighbors = avg(getArrayNeo4j('kneighbors1'))/1000
wiki_neo4j_sp = avg(getArrayNeo4j('sp1'))/1000
pokec_neo4j_sum = avg(getArrayNeo4j('sum2'))/1000
pokec_neo4j_avg = avg(getArrayNeo4j('avg2'))/1000
pokec_neo4j_count = avg(getArrayNeo4j('count2'))/1000
pokec_neo4j_min = avg(getArrayNeo4j('min2'))/1000
pokec_neo4j_max = avg(getArrayNeo4j('max2'))/1000
pokec_neo4j_kneighbors = avg(getArrayNeo4j('kneighbors2'))/1000
pokec_neo4j_sp = avg(getArrayNeo4j('sp2'))/1000
```

```python
labels = ['Sum', 'Avg', 'Count', 'Min', 'Max']
wiki_arango_means = [wiki_arango_sum, wiki_arango_avg, wiki_arango_count, wiki_arango_min, wiki_arango_max]
wiki_neo4j_means = [wiki_neo4j_sum, wiki_neo4j_avg, wiki_neo4j_count, wiki_neo4j_min, wiki_neo4j_max]

wiki_arango_std = divideLists(wiki_arango_means, wiki_arango_means)
wiki_neo4j_std = divideLists(wiki_neo4j_means, wiki_arango_means)

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, wiki_arango_std, width, label='Arango')
rects2 = ax.bar(x + width/2, wiki_neo4j_std, width, label='Neo4j')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percentage of ArangoDB Time')
ax.set_title('Summarization Queries Benchmarks for Wiki Talk Temporal')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(loc='center right')

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(str(round(height))+'%'),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
